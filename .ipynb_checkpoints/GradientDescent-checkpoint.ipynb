```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 3, 4, 5,2])
Y = np.array([3, 4, 2, 5, 6,2])
```

```python
m,c=0,0
```

```python
epochs=1000
L=0.0001
```

```python
for i in range(epochs):
    y_pred=m*X+c
    dm=(-2/len(X))*sum(X*(Y-y_pred))
    dc=(-2/len(X))*sum(Y-y_pred)
    m-=dm*L
    c-=dc*L
    if(i%100==0):
        print(f"slope:{m} intercept: {c}")
        loss = np.mean((Y - y_pred) ** 2)
        print(f"loss: {loss}")
```

```python
plt.scatter(X, Y, color='blue', label='Actual Points')
plt.plot(X, m*X + c, color='red', label='Fitted Line')
plt.legend()
plt.title("Linear Fit using Gradient Descent")
plt.show()
```
