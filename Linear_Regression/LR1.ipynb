```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r'house_prediction.csv')
```

```python
# dataset.drop(['yr_built'], axis=1, inplace=True)
# print(dataset.head())
dataset.head()
```

```python
# divide into independent and dependent features
```

```python
# X=dataset.iloc[:,2:]
X = dataset.drop(dataset.columns[0], axis=1)  # drop the target column
```

```python
X.head()
```

```python
Y=dataset['price']
```

```python
Y.head()
```

```python
from sklearn.linear_model import LinearRegression
# do lot of train and test splits size varies 
from sklearn.model_selection import cross_val_score
```

```python
type(X)
print(type(Y))
print(type(X))
```

```python
X.head()
```

```python
Y.head()
```

```python
lin_reg=LinearRegression()
mse = cross_val_score(lin_reg, X, Y, scoring='neg_mean_squared_error', cv=5)
print(mse)
```

```python
lin_reg.fit(X, Y)
```

```python
avg_mse=np.mean(mse)
print(avg_mse)
```

```python
sample = pd.DataFrame([[2.0,1.5, 1250, 8000, 2, 1500, 500]],
                      columns=['bedrooms','bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement'])
```

```python
predicted_price = lin_reg.predict(sample)
print("Predicted Price:", predicted_price[0])
```
